```python
import numpy as np
def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    nwords = 0.
    
    for word in words:
        if word in vocabulary:
            nwords = nwords + 1
            feature_vector = np.add(feature_vector, model.wv[word])
    
    if nwords:
        feature_vector = np.divide(feature_vector, nwords)
    
    return feature_vector
preprocessed_reviews = data_final["stemmed_review"]
document_vectors = [average_word_vectors(review, model, model.wv.index_to_key, 100) for review in preprocessed_reviews]
```

```python
import string
from nltk.stem.porter import PorterStemmer
exclude = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("","", exclude))

def remove_stop_words(text):
    new_text =[]
    stop_words = set(stopwords.words('english'))
    for word in text.split():
        if word in stop_words:
            new_text.append("")
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return" ".join(x)
def remove_emoji(text):
    new_text = emoji.demojize(text)
    return new_text

def stem_word(text):
    ps = PorterStemmer()
    stemmed_words = [ps.stem(word) for word in text]
    return stemmed_words
```

```python
from gensim.models import Word2Vec
import numpy as np

# Assuming word2vec_model is your trained Word2Vec model
def prediction(text):
    # Preprocess the text
    re_pun_text = remove_punctuation(text)
    re_stop_text = remove_stop_words(re_pun_text)
    re_emoji = remove_emoji(re_stop_text)
    all_re_text = re_emoji.lower()
    token_text = word_tokenize(all_re_text)
    stemmed = stem_word(token_text)
    
    # Convert stemmed words to Word2Vec embeddings
    word_embeddings = []
    for word in stemmed:
        if word in model.wv:
            word_embeddings.append(model.wv[word])
        else:
            word_embeddings.append(np.zeros(model.vector_size))  # Use zero vector for unknown words
    
    # Calculate the mean of word embeddings as document vector
    if word_embeddings:
        doc_vector = np.mean(word_embeddings, axis=0)
    else:
        doc_vector = np.zeros(model.vector_size)  # Use zero vector for empty input
    
    # Reshape the document vector for prediction
    doc_vector_reshaped = doc_vector.reshape(1, -1)

    # Make the prediction
    prediction = clf.predict(doc_vector_reshaped)[0]
    return prediction

```
